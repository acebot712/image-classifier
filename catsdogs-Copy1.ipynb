```python
import os
import zipfile

def extract_zip_file(zip_file, extract_to):
    """
    Extract a zip file to a specified directory.
    """
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

# Define input and extraction directories
local_zip = 'dogs-vs-cats.zip'
extract_dir = './tmp'

try:
    # Extract the main zip file
    extract_zip_file(local_zip, extract_dir)
    
    # Loop through sub-folders and extract their contents
    for root, dirs, files in os.walk(extract_dir):
        for file in files:
            if file.endswith('.zip'):
                sub_zip_path = os.path.join(root, file)
                extract_zip_file(sub_zip_path, root)
                os.remove(sub_zip_path)  # Remove the sub-zip file
                
    print("Extraction completed.")
    
except Exception as e:
    print("An error occurred:", e)
```

```python
# Import the necessary dependencies
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
from PIL import Image
import seaborn as sns
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
```

```python
def load_and_preprocess_data(data_dir, categories, img_size):
    """
    Load and preprocess image data from the specified directory.
    """
    images = []
    labels = []

    for img_file in os.listdir(data_dir):
        img_path = os.path.join(data_dir, img_file)
        img = Image.open(img_path).convert('L')
        img = img.resize((img_size, img_size))
        img = np.array(img, dtype=np.float32) / 255.0
        label = img_file.split(".")[0]
        label_index = categories.index(label)
        images.append(img)
        labels.append(label_index)

    # Convert the lists to NumPy arrays
    images = np.array(images)
    labels = np.array(labels)
    
    return images, labels

data_dir = 'tmp/train'
categories = ['dog', 'cat']
img_size = 100

try:
    images, labels = load_and_preprocess_data(data_dir, categories, img_size)
    print("Data loading and preprocessing completed.")
    
    # Proceed with using 'images' and 'labels' for training or analysis
    
except Exception as e:
    print("An error occurred:", e)
```

```python
# Step 2: Define dataset class
class CustomDataset(Dataset):
    """
    Custom dataset class for image data.
    """

    def __init__(self, images, labels=None, transform=None):
        """
        Initialize the dataset.

        Args:
            images (numpy.ndarray): Array of images.
            labels (numpy.ndarray, optional): Array of labels. Default is None.
            transform (callable, optional): Optional transform to be applied to images. Default is None.
        """
        self.images = images
        self.labels = labels
        self.transform = transform

    def __len__(self):
        """
        Get the number of samples in the dataset.

        Returns:
            int: Number of samples.
        """
        return len(self.images)

    def __getitem__(self, idx):
        """
        Get a sample from the dataset.

        Args:
            idx (int): Index of the sample.

        Returns:
            tuple: A tuple containing the image and label (if available).
        """
        img = self.images[idx]
        label = self.labels[idx] if self.labels is not None else -1

        if self.transform:
            img = self.transform(img)

        return img, label

```

```python
# Step 3: Preprocessing and Data Loaders
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])
```

```python
# Step 4: Create the dataset using the CustomDataset class
dataset = CustomDataset(images, labels, transform=transform)
batch_size = 32
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
# Step 5: Create a Convolutional Neural Network
class Net(nn.Module):
    """
    Custom Convolutional Neural Network class.
    """

    def __init__(self, num_classes, img_size):
        """
        Initialize the CNN.

        Args:
            num_classes (int): Number of output classes.
            img_size (int): Size of input images (assuming square images).
        """
        super().__init__()

        self.features = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        self.flatten = nn.Flatten()

        self.classifier = nn.Sequential(
            nn.Linear(32 * (img_size // 4) ** 2, 128),
            nn.ReLU(inplace=True),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        """
        Forward pass of the network.

        Args:
            x (torch.Tensor): Input tensor.

        Returns:
            torch.Tensor: Output tensor.
        """
        x = self.features(x)
        x = self.flatten(x)
        x = self.classifier(x)
        return x
```

```python
# Step 6: Training loop
device = torch.device("cuda" if torch.cuda.is_available() else "mps" if torch.backends.mps.is_available() else "cpu")
num_classes = len(categories)
img_size = 100
model = Net(num_classes, img_size).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

num_epochs = 10
```

```python
def train_model(model, dataloader, criterion, optimizer, num_epochs, device):
    """
    Train the model using the provided data and configurations.

    Args:
        model (nn.Module): The neural network model.
        dataloader (DataLoader): The data loader.
        criterion: Loss function.
        optimizer: Optimization algorithm.
        num_epochs (int): Number of training epochs.
        device: Device to use for training (e.g., 'cuda' or 'cpu').

    Returns:
        None
    """
    for epoch in range(num_epochs):
        model.train()
        for images, labels in dataloader:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
        
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")

# Train the model
train_model(model, dataloader, criterion, optimizer, num_epochs, device)
    
```

```python
# Save the model after each epoch
model_save_path = f'model_epoch_{epoch+1}.pth'
torch.save(model.state_dict(), model_save_path)
print(f"Saved model checkpoint: {model_save_path}")
```

```python
def load_and_preprocess_image(image_path, img_size):
    """
    Load and preprocess a single image.

    Args:
        image_path (str): Path to the image file.
        img_size (int): Size of the image (assuming square images).

    Returns:
        torch.Tensor: Preprocessed image tensor.
    """
    img = Image.open(image_path).convert('L')
    img = img.resize((img_size, img_size))
    img = np.array(img, dtype=np.float32) / 255.0
    
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5], std=[0.5])
    ])
    img_tensor = transform(img).unsqueeze(0)  # Add batch dimension
    
    return img_tensor

# Load the trained model
model = Net(num_classes=len(categories), img_size=100)
model.load_state_dict(torch.load('model_epoch_10.pth'))
model.eval()

# Preprocess the random image
random_image_path = 'tmp/train/dog.8011.jpg'
img_tensor = load_and_preprocess_image(random_image_path, img_size)

# Perform inference
with torch.no_grad():
    output = model(img_tensor)

# Get the predicted class
predicted_class = torch.argmax(output, dim=1).item()
predicted_label = categories[predicted_class]

print(f"Predicted Class: {predicted_class} - {predicted_label}")
```

```python
class CustomDatasetTest(Dataset):
    def __init__(self, images, transform=None):
        self.images = images
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img = self.images[idx]
        if self.transform:
            img = self.transform(img)
        return img
        
def load_and_preprocess_test_images(test_data_dir, img_size, transform=None):
    """
    Load and preprocess test images from the specified directory.

    Args:
        test_data_dir (str): Directory containing test images.
        img_size (int): Size of images (assuming square images).
        transform (callable, optional): Optional transform to be applied to images. Default is None.

    Returns:
        np.ndarray: Preprocessed test images.
    """
    test_images = []
    for img_file in os.listdir(test_data_dir):
        img_path = os.path.join(test_data_dir, img_file)
        img = Image.open(img_path).convert('L')
        img = img.resize((img_size, img_size))
        img = np.array(img, dtype=np.float32) / 255.0
        if transform:
            img = transform(img)
        test_images.append(img)
    return np.array(test_images)

# Step 1: Load the trained model
model = Net(num_classes=len(categories), img_size=100)
model.load_state_dict(torch.load('model_epoch_10.pth'))
model.eval()

# Step 2: Load and preprocess the test data
test_data_dir = 'tmp/test1'
test_images = load_and_preprocess_test_images(test_data_dir, img_size, transform=transform)

# Step 3: Create a DataLoader for the test set
test_dataset = CustomDatasetTest(test_images)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
# Generate predictions for submission
device = torch.device("cuda" if torch.cuda.is_available() else "mps" if torch.backends.mps.is_available() else "cpu")
model.to(device)

def generate_predictions(model, dataloader, device):
    """
    Generate predictions using the model and dataloader.

    Args:
        model (nn.Module): The neural network model.
        dataloader (DataLoader): DataLoader for test data.
        device: Device to use for inference (e.g., 'cuda' or 'cpu').

    Returns:
        list: List of predictions.
    """
    all_predictions = []
    model.eval()  # Set model to evaluation mode
    with torch.no_grad():
        for images in dataloader:
            images = images.to(device)
            outputs = model(images)
            predictions = torch.argmax(outputs, dim=1)
            all_predictions.extend(predictions.cpu().numpy())
    return all_predictions

# Step 4: Generate predictions for submission
predictions = generate_predictions(model, test_dataloader, device)
```

```python
# Generate submission CSV
submission_df = pd.DataFrame({'id': np.arange(1, len(predictions) + 1), 'label': predictions})
submission_df.to_csv('submission.csv', index=False)

print("Submission CSV generated.")
```
