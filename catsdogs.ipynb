```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import torch
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torchvision.transforms as transforms
```

```python
# Step 2: Load data and preprocess
data_dir = 'train'
categories = ['dog', 'cat']

img_size = 100
images = []
labels = []

for img_file in os.listdir(data_dir):
    img_path = os.path.join(data_dir, img_file)
    img = Image.open(img_path).convert('L')
    img = img.resize((img_size, img_size))
    img = np.array(img, dtype=np.float32) / 255.0
    label = img_file.split(".")[0]
    label_index = categories.index(label)
    images.append(img)
    labels.append(label_index)

np.random.shuffle(images)
np.random.shuffle(labels)

# Convert the lists to NumPy arrays
images = np.array(images)
labels = np.array(labels)
```

```python
# Step 1: Define dataset class
class CustomDataset(Dataset):
    def __init__(self, images, labels, transform=None):
        self.images = images
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img = self.images[idx]
        label = self.labels[idx]
        if self.transform:
            img = self.transform(img)
        return img, label
```

```python
# Step 4: Preprocessing and Data Loaders
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

```

```python
df.values[0].dtype
```

```python
# Create the dataset using the CustomDataset class
dataset = CustomDataset(images, labels, transform=transform)
batch_size = 32
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
class Net(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(16,32, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2) # 25 * 25 *32
        )

        self.flatten = nn.Flatten()

        self.classifier = nn.Sequential(
            nn.Linear(32 * (img_size // 4) ** 2, 128),
            nn.ReLU(inplace=True),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.flatten(x)
        x = self.classifier(x)
        return x
        
```

```python
# Step 6: Training loop
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = Net(len(categories)).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

num_epochs = 10
```

```python
for epoch in range(num_epochs):
    model.train()
    for images, labels in dataloader:
        images, labels = images.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
    
```

```python
# Save the model after each epoch
model_save_path = f'model_epoch_{epoch+1}.pth'
torch.save(model.state_dict(), model_save_path)
print(f"Saved model checkpoint: {model_save_path}")
```

```python
# Load the trained model
model = Net(num_classes=len(categories))
model.load_state_dict(torch.load('model_epoch_10.pth'))
model.eval()

# Preprocess the random image
random_image_path = 'train/dog.8011.jpg'
img = Image.open(random_image_path).convert('L')
img = img.resize((img_size, img_size))
img = np.array(img, dtype=np.float32) / 255.0
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])
img_tensor = transform(img).unsqueeze(0)  # Add batch dimension

# Perform inference
with torch.no_grad():
    output = model(img_tensor)

# Get the predicted class
predicted_class = torch.argmax(output, dim=1).item()
predicted_label = categories[predicted_class]

print(f"Predicted Class: {predicted_class} - {predicted_label}")
```
